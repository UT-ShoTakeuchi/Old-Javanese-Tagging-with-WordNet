Now we can try some translation.

We will make a OJ-English dictionary from the WordNet, and see if we have translation for each word in  a prose.

The "prose" object below are sentences from Kawi prose shown in p.15 of Zurbuchen (1976) . (See READ ME)

```python
from nltk.corpus import wordnet as wn

wdic = {}
errors = []

for line in open("wn-kaw_edit.tab"):
  line = line.strip()
  lnls = line.split('\t')
  posoff = lnls[0].split('-')
  en_syn = wn.synset_from_pos_and_offset(posoff[1], int(posoff[0]))
  en_lemma = en_syn.lemmas()
  en_corrword = en_lemma[0].name()
  en_syn_def = en_syn.definition()
  wdic[lnls[1]] = en_corrword

prose = 'Inĕh nireng ratha ikang kanya., I pĕjah sang Hidimba, manembah ta sang Hidimbi ri Dewi Kunti. Kunang ikang agĕni haneng kunda sadākala murub. Yapwan sang umayat ikang laras yekā sang Arjuna ngaranya. Inarpane sang brāhmana lĕmbu satus de Mahārāja Janamejaya. Ana sira ratu ulun paṇḍita sang Sewāntara ngaranira.'

vl = prose.split()
for i in vl:
  wd = i.strip(r'[\,\.\(\)]')
  wd = wd.lower()
  if wd in wdic:
    print(wd+'\t'+wdic[wd])
  else:
    print(wd+'\t'+'NO TRANSLATION AVAILABLE')
```
